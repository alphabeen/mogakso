```python
# 1. ÎùºÏù¥Î∏åÎü¨Î¶¨ Î°úÎî©
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 2. Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎî© (Titanic ÎòêÎäî ÏÇ¨Ïö©Ïûê CSV)
df = sns.load_dataset('titanic')

# 3. Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨: Í≤∞Ï∏°Ïπò Ï†úÍ±∞
df_clean = df[['age', 'fare', 'survived']].dropna()

# 4. ÏÉÅÍ¥ÄÍ≥ÑÏàò Î∂ÑÏÑù (Pearson ÏÉÅÍ¥ÄÍ≥ÑÏàò)
correlation_matrix = df_clean.corr()
print("üìå ÏÉÅÍ¥ÄÍ≥ÑÏàò:\n", correlation_matrix)

# 5. Heatmap ÏãúÍ∞ÅÌôî
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# 6. ÌöåÍ∑Ä Î∂ÑÏÑù: ÎÇòÏù¥ vs ÏöîÍ∏à (fare)
sns.lmplot(x='age', y='fare', data=df_clean, height=5, aspect=1.5)
plt.title('Age vs Fare: Regression Line')
plt.show()

# 7. ÌöåÍ∑Ä Î∂ÑÏÑù: ÎÇòÏù¥ÏôÄ ÏÉùÏ°¥ Ïó¨Î∂Ä
sns.lmplot(x='age', y='survived', data=df_clean, logistic=True)
plt.title('Age vs Survival (Logistic Regression View)')
plt.show()
```
