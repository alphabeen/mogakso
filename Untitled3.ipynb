```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 예제 데이터셋 로드 (타이타닉 또는 CSV 파일 등)
df = sns.load_dataset('titanic')  # 또는 pd.read_csv('your_file.csv')

# 1. 기술 통계 지표 계산
ages = df['age'].dropna()
mean_age = ages.mean()
median_age = ages.median()
std_age = ages.std()

print(f"평균 나이: {mean_age:.2f}")
print(f"중앙값 나이: {median_age:.2f}")
print(f"표준편차: {std_age:.2f}")

# 2. 간단한 기술 통계 시각화 (히스토그램 + 평균선)
plt.figure(figsize=(8, 5))
sns.histplot(ages, bins=30, kde=True)
plt.axvline(mean_age, color='red', linestyle='--', label='Mean')
plt.axvline(median_age, color='green', linestyle='--', label='Median')
plt.title('Titanic Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()

# 3. EDA: 나이와 생존 여부의 관계
plt.figure(figsize=(8, 5))
sns.boxplot(x='survived', y='age', data=df)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.tight_layout()
plt.show()

# 4. 생존율 통계 요약
survival_rate = df['survived'].mean()
print(f"전체 생존율: {survival_rate:.2%}")
```
